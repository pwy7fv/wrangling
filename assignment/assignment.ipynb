```python
ds = pd.read_csv('wrangling/data/sharks.csv', low_memory=False)
ds.head()
```

```python
type = ds['Type']
type.unique()
```

```python
type.value_counts()
```

```python
type = type.replace(['Watercraft','Boat','Boating','Boatomg'], 'Watercraft')
type.value_counts() # condensed these down into one variable cause they are essentially the same just worded differently
```

```python
SentenceLength = dp['ImposedSentenceAllChargeInContactEvent']
print(SentenceLength.unique(), '\n')
SentenceType = dp['SentenceTypeAllChargesAtConvictionInContactEvent']
print(SentenceType.unique(), '\n')
```

```python
SentenceLength = pd.to_numeric(SentenceLength,errors='coerce') #changing to numeric, easier to work with
SentenceLength_nan = SentenceLength.isnull() # seeing if data is missing or not with 1 or 0

print('After coercion: \n', SentenceLength.describe(),'\n')
print('Total Missings: \n', sum(SentenceLength_nan),'\n')
```

```python
print(pd.crosstab(SentenceLength_nan, SentenceType))
```

```python
SentenceLength = SentenceLength.mask(SentenceType == 4,0) #Based on the crosstab you can see that when there is a
                                    # 4(case dismissed) then there is no sentence and so no length of a sentence
SentenceLength = SentenceLength.mask(SentenceType == 9,np.nan) #9 is actually missing data so should just be np.nan

SentenceLength_nan = SentenceLength.isnull() # Create a new missing dummy
print( pd.crosstab(SentenceLength_nan, SentenceType), '\n')
print('Total Missings: \n', sum(SentenceLength_nan),'\n')

dp['ImposedSentenceAllChargeInContactEvent'] = SentenceLength
```
